Cargar los datos

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset
df = pd.read_csv("data/kc_house_data.csv")
df.head()
```

Tema: Distribución del número de dormitorios en las propiedades


```python
import matplotlib.pyplot as plt

# Contar la cantidad de propiedades por número de dormitorios
bedroom_distribution = df['bedrooms'].value_counts().sort_index()

plt.figure(figsize=(8, 8))
plt.pie(bedroom_distribution, labels=bedroom_distribution.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribución del Número de Dormitorios en las Propiedades')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is a circle.
plt.show()
```

Este gráfico circular muestra la proporción de propiedades según el número de dormitorios. Cada porción representa un número específico de dormitorios (por ejemplo, 1 dormitorio, 2 dormitorios, etc.), permitiendo visualizar fácilmente cual es el porcentaje de propiedades que hay en cada categoría.
