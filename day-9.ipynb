Cargar los datos

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset
df = pd.read_csv("data/kc_house_data.csv")
df.head()
```

Tema: Comparación de precios promedio por número de dormitorios


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calcular el precio promedio por número de dormitorios
avg_price_by_bedrooms = df.groupby('bedrooms')['price'].mean().sort_index()

plt.figure(figsize=(10, 6))
sns.barplot(x=avg_price_by_bedrooms.index, y=avg_price_by_bedrooms.values, palette='viridis')
plt.title('Precio Promedio de Propiedades por Número de Dormitorios')
plt.xlabel('Número de Dormitorios')
plt.ylabel('Precio Promedio ($)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

Este gráfico de barras muestra el precio promedio de las propiedades según el número de dormitorios. Cada barra representa el precio promedio correspondiente a un número específico de dormitorios, permitiendo realizar comparaciones claras entre las diferentes categorías.
