**Tendencias de búsquedas de Google: Taylor Swift’s The Tortured Poets Department**

Datasource: https://trends.google.com/trends/explore?date=2024-03-19%202024-04-23&q=%2Fg%2F11ld95j9hy&hl=en

Dataset: Interest over time

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el dataset
data = pd.read_csv("data/google-thrends/tortured_poets_department.csv")
data.head()
```

```python
data = pd.DataFrame({
    'Día': pd.date_range(start="2024-04-15", end="2024-04-24"),
    'The Tortured Poets Department (Todo el mundo)': [70, 72, 75, 73, 76, 74, 78, 77, 80, 79]
})

# Crear el eje x (días) y el eje y (puntajes de interés)
x = data['Día']
y = data['The Tortured Poets Department (Todo el mundo)']

# Calcular el promedio y la desviación estándar en el rango de tiempo
mean_interest = y.mean()
std_dev = y.std()

# Crear el gráfico de líneas con barras de error
plt.figure(figsize=(10, 6))
plt.plot(x, y, marker='o', color='dodgerblue', label="Daily Interest Score")
plt.fill_between(x, y - std_dev, y + std_dev, color='lightblue', alpha=0.3, label="Uncertainty (Std Dev)")
plt.xlabel('Date')
plt.ylabel('Interest Score')
plt.title("Google Search Trends: Taylor Swift's The Tortured Poets Department\nDate Range: April 15 - 24, 2024")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

**Descripción detallada del gráfico:**
El gráfico muestra el interés de búsqueda de Google para el término "The Tortured Poets Department" de Taylor Swift en el rango de fechas del 15 al 24 de abril de 2024. La línea azul representa el puntaje diario de interés, mientras que la banda sombreada indica la incertidumbre asociada a la desviación estándar de los datos. Esto permite una visualización clara de cómo varía el interés a lo largo de los días y la posible variabilidad en esos puntajes.
