Cargar los datos

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset
df = pd.read_csv("data/Predict_student_dropout_and_academic_success.csv", sep=";")
df.head()
```

Tema: RelaciÃ³n entre la calificaciÃ³n de la madre y el estado de la matrÃ­cula de los estudiantes

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Mother\'s qualification', hue='Tuition fees up to date', data=df, palette='Set2')
plt.title('Impacto de la CalificaciÃ³n de la Madre en el Estado de la MatrÃ­cula')
plt.xlabel('CalificaciÃ³n de la Madre')
plt.ylabel('NÃºmero de Estudiantes')
plt.xticks(rotation=45)
plt.legend(title='MatrÃ­cula al DÃ­a', loc='upper right', labels=['SÃ­', 'No'])
plt.grid(axis='y')
plt.show()
```

Este grÃ¡fico de barras muestra la distribuciÃ³n del estado de matrÃ­cula (si estÃ¡ al dÃ­a o no) en funciÃ³n de la calificaciÃ³n de la madre. Cada barra representa la cantidad de estudiantes con una calificaciÃ³n materna especÃ­fica, diferenciando entre aquellos cuya matrÃ­cula estÃ¡ al dÃ­a y aquellos que no lo estÃ¡
