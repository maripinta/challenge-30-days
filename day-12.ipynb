Cargar los datos

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset
df = pd.read_csv("data/Predict_student_dropout_and_academic_success.csv", sep=";")
df.head()
```

Tema: Relación entre la calificación de la madre y el estado de la matrícula de los estudiantes

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Mother\'s qualification', hue='Tuition fees up to date', data=df, palette='Set2')
plt.title('Impacto de la Calificación de la Madre en el Estado de la Matrícula')
plt.xlabel('Calificación de la Madre')
plt.ylabel('Número de Estudiantes')
plt.xticks(rotation=45)
plt.legend(title='Matrícula al Día', loc='upper right', labels=['Sí', 'No'])
plt.grid(axis='y')
plt.show()
```

Este gráfico de barras muestra la distribución del estado de matrícula (si está al día o no) en función de la calificación de la madre. Cada barra representa la cantidad de estudiantes con una calificación materna específica, diferenciando entre aquellos cuya matrícula está al día y aquellos que no lo está
