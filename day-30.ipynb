Dataset: alcohol-consumption

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el dataset
data = pd.read_csv("data/five-thirty-eight-datssets/drinks.csv")
data.head()
```

```python
# Calcular el promedio y la desviación estándar para cada tipo de bebida
mean_values = data[['beer_servings', 'spirit_servings', 'wine_servings']].mean()
std_dev = data[['beer_servings', 'spirit_servings', 'wine_servings']].std()

# Crear el gráfico de barras con barras de error (incertidumbre)
plt.figure(figsize=(10, 6))
plt.bar(mean_values.index, mean_values, yerr=std_dev, color=['#ffa07a', '#20b2aa', '#9370db'], capsize=10)
plt.xlabel('Tipo de alcohol')
plt.ylabel('Porciones promedio por país')
plt.title('Consumo promedio de alcohol con incertidumbre por tipo')
plt.tight_layout()
plt.show()
```

**Descripción detallada del gráfico**: El gráfico muestra el consumo promedio de tres tipos de bebidas alcohólicas (cerveza, licores y vino) en distintos países, con barras de error que indican la desviación estándar para cada tipo de bebida. Esto refleja la variabilidad en el consumo promedio entre los países, indicando la dispersión de los datos en cada categoría de bebida.

Personas interesadas en patrones de consumo de alcohol por país, investigadores de salud pública, sociólogos, y responsables de políticas de salud que buscan entender las tendencias y diferencias de consumo en distintas regiones.

**Objetivo**:
El objetivo es proporcionar una visión general de los hábitos de consumo de alcohol promedio por tipo de bebida y mostrar la variabilidad (incertidumbre) en cada categoría para entender las diferencias de consumo a nivel mundial.

**Mensaje**:
Este gráfico enfatiza que aunque los promedios de consumo por tipo de bebida pueden ser similares entre países, existe una significativa variabilidad que puede estar influenciada por factores culturales, económicos y de políticas de cada región.
