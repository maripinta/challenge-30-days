Cargar los datos

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset
df = pd.read_csv("data/Predict_student_dropout_and_academic_success.csv", sep=";")
df.head()
```

Tema: Distribución de la edad al momento de la inscripción de los estudiantes

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df['Age at enrollment'], bins=10, kde=True, color='skyblue')
plt.title('Distribución de la Edad al Momento de la Inscripción')
plt.xlabel('Edad al Inscribirse')
plt.ylabel('Número de Estudiantes')
plt.grid(axis='y')
plt.show()
```

Este gráfico de histogramas muestra la distribución de la edad de los estudiantes en el momento de su inscripción. La curva de densidad (KDE) ayuda a visualizar la tendencia general de las edades de los estudiantes.
