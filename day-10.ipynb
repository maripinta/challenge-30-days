Cargar los datos

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset
df = pd.read_csv("data/Predict_student_dropout_and_academic_success.csv", sep=";")
df.head()
```

Tema: DistribuciÃ³n de la edad al momento de la inscripciÃ³n de los estudiantes

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df['Age at enrollment'], bins=10, kde=True, color='skyblue')
plt.title('DistribuciÃ³n de la Edad al Momento de la InscripciÃ³n')
plt.xlabel('Edad al Inscribirse')
plt.ylabel('NÃºmero de Estudiantes')
plt.grid(axis='y')
plt.show()
```

Este grÃ¡fico de histogramas muestra la distribuciÃ³n de la edad de los estudiantes en el momento de su inscripciÃ³n. La curva de densidad (KDE) ayuda a visualizar la tendencia general de las edades de los estudiantes.
