Cargar los datos

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset
df = pd.read_csv("data/kc_house_data.csv")
df.head()
```

Tema: Desviación del precio promedio de venta según la zona



```python
avg_price_by_zipcode = df.groupby("zipcode")["price"].mean()
overall_mean_price = df["price"].mean()
price_diff = avg_price_by_zipcode - overall_mean_price

# Ordenar por la desviación del promedio
price_diff = price_diff.sort_values()

plt.figure(figsize=(12, 8))
plt.barh(price_diff.index, price_diff, color=['#FF595E' if x < 0 else '#1982C4' for x in price_diff])
plt.axvline(x=0, color='gray', linestyle='--')
plt.title("Diferencias en el precio promedio de venta por zona postal")
plt.xlabel("Desviación del precio promedio ($)")
plt.ylabel("Zona postal")
plt.show()
```

Este gráfico muestra cómo el precio promedio de venta en cada zona postal se desvía del precio promedio general. Las barras hacia la derecha indican zonas donde los precios superan el promedio, mientras que las barras hacia la izquierda representan zonas más asequibles. Este enfoque destaca claramente las áreas más exclusivas y las más económicas.
