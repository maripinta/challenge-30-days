Brecha salarial de género
La brecha salarial de género se define como la diferencia entre los ingresos medios de hombres y mujeres en relación con los ingresos medios de los hombres.
Datasource: https://www.oecd.org/en/data/indicators/gender-wage-gap.html

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset
data = pd.read_csv("data/deco/gender_wage_gap.csv")
data.head()

```

Tema: Brecha salarial de género

Porcentaje de los ingresos medios de los hombres, 2022

```python
# Crear el gráfico
plt.figure(figsize=(20, 6))
sns.barplot(data=data, x='Category', y='Gender wage gap', palette='pastel')

# Personalizar el gráfico
plt.title("Gender Wage Gap Across Sectors in 2022")
plt.xlabel("Sector")
plt.ylabel("Gender Wage Gap (%)")
plt.xticks(rotation=45)
plt.show()
```

**Descripción detallada:**
Este gráfico de barras compara la brecha salarial de género en diferentes sectores económicos en 2022. Cada barra representa el porcentaje de la brecha salarial en un sector, con valores que indican la proporción en la que el salario de las mujeres se sitúa por debajo del salario medio de los hombres en cada categoría.

**Audiencia:**
Este gráfico está dirigido a investigadores, responsables de políticas de igualdad, organizaciones de derechos laborales, y profesionales de recursos humanos interesados en la equidad salarial.

**Objetivo:**
Visualizar las diferencias en la brecha salarial de género por sector en 2022, proporcionando una referencia clara de los sectores donde esta disparidad es más y menos pronunciada.

**Mensaje:**
La visualización destaca los sectores con mayores y menores diferencias salariales de género, mostrando claramente la magnitud de la brecha en cada uno.

**Tipo de visualización y Justificación de su uso:**
Gráfico de barras de comparación, que permite una comparación directa entre diferentes categorías o sectores.

Un gráfico de barras facilita la comparación de la brecha salarial entre sectores, ya que presenta los datos de manera directa y fácil de interpretar. Este tipo de gráfico permite observar rápidamente qué sectores tienen mayores disparidades salariales, ayudando a identificar áreas clave donde es necesario reducir la brecha de género en los ingresos.
