## Día 3: Makeover

### Tema: Visualización mejorada de precios según el número de habitaciones

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.boxplot(x='bedrooms', y='price', data=df)
plt.title("Comparación de precios por número de habitaciones")
plt.xlabel("Número de habitaciones")
plt.ylabel("Precio ($)")
plt.show()
```

El gráfico muestra la dispersión y variación del precio de venta de propiedades según la cantidad de habitaciones. Cada recuento de habitaciones (ejemplo: 1, 2, 3, etc.) tiene un boxplot que visualiza el rango intercuartil, el valor medio, y los valores atípicos. Esto permite ver cómo aumentan los precios con el número de habitaciones y observar la gran dispersión de precios entre las propiedades de mayor tamaño.
