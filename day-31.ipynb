Cargar los datos

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset
df = pd.read_csv("data/kc_house_data.csv")
df.head()
```

Tema: Relación entre el precio y el tamaño (en pies cuadrados) de las propiedades


```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='sqft_living', y='price', alpha=0.5)
plt.title('Relación entre Precio y Tamaño de la Propiedad')
plt.xlabel('Tamaño (pies cuadrados)')
plt.ylabel('Precio ($)')
plt.grid(True)
plt.show()
```

Este gráfico de dispersión muestra la relación entre el tamaño de las propiedades (en pies cuadrados) y su precio de venta. Cada punto representa una propiedad, lo que permite observar la tendencia general de que a mayor tamaño, mayor es el precio, aunque también se pueden identificar propiedades que tienen un precio elevado a pesar de ser más pequeñas.
