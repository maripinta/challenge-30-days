Cargar los datos

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset
df = pd.read_csv("data/Predict_student_dropout_and_academic_success.csv", sep=";")
df.head()
```

Tema: ComparaciÃ³n de ocupaciones de los padres de los estudiantes

```python
# Contar las ocupaciones de los padres
father_occupation = df['Father\'s occupation'].value_counts().nlargest(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=father_occupation.values, y=father_occupation.index, palette='pastel')
plt.title('Top 10 Ocupaciones de los Padres de los Estudiantes')
plt.xlabel('NÃºmero de Estudiantes')
plt.ylabel('OcupaciÃ³n del Padre')
plt.grid(axis='x')
plt.show()


```

Este grÃ¡fico de barras muestra las diez ocupaciones mÃ¡s comunes de los padres de los estudiantes. Cada barra representa el nÃºmero de estudiantes cuyo padre tiene una ocupaciÃ³n especÃ­fica.
